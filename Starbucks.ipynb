```python
import matplotlib.pyplot as plt  
import pandas as pd  
#%matplotlib inline
'exec(%matplotlib inline)'
 
import numpy as np  

 
print ('\ncustomer data\n')
popular_time=np.array([1,1,2,2,3,3,3,3,4,4,3,5,7,5,4,6,7,7,8,8])
hot_beverages=np.array([5,7,6,9,6,8,8,9,8,8,3,4,2,7,5,9,3,9,1,7])
point = ['P1','P2','P3','P4','P5','P6','P7','P8','P9','P10','P11','P12','P13','P14','P15','P16','P17','P18','P19','P20']
data = pd.DataFrame({'Point':point, 'popular_time':np.round(popular_time,2), 'hot_beverages':np.round(hot_beverages,2)})
data = data.set_index('Point')
data
```

Average Link

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Customer Dendograms")  
dend = shc.dendrogram(shc.linkage(data, method='average'))  
print ('\npring dend \n')
print(dend)
plt.show() 
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='manhattan', 
linkage='average')  
print(cluster.fit_predict(data))
plt.figure(figsize=(10, 7))  
plt.scatter(data['popular_time'], data['hot_beverages'], c=cluster.labels_, cmap='rainbow')  
plt.show()
```

```python
beverages={1:"coffee Type", 2:"cafe latte",3:"cafe mocha",4:"cappucino",5:"caramel macchiato",6:"cafe americano",7:"hot chocolate",8:"espresso shot",9:"coffeeof the day"}
```

```python
timings={1:"6am - 7 am", 2:"7am - 8 am",3:"8 am - 9 am",4:"9am - 10am",5:"10am - 12 noon",6:"12:00 - 1:00 pm",7:"1:00 pm - 3:00 pm",8:"3:00 pm - 4:00 pm",9:"4:00 pm - 7:00 pm",10:"7:00 pm - 8:00 pm",11:"8:00 pm - 9:00 pm"}
```

```python
data['timings']=data['popular_time'].map(timings)
```

```python
data['beverages']=data['hot_beverages'].map(beverages)
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
x1=np.array([1,1,2,2,3,3,3,3,4,4,3,5,7,5,4,6,7,7,8,8])
x2=np.array([5,7,6,9,6,8,8,9,8,8,3,4,2,7,5,9,3,9,1,7])
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
colors = ['b', 'g', 'c']
markers = ['o', 'v', 's']

# KMeans algorithm 
K = 3
kmeans_model = KMeans(n_clusters=K).fit(X)

print(kmeans_model.cluster_centers_)
centers = np.array(kmeans_model.cluster_centers_)

plt.plot()
plt.title('k means centroids')

for i, l in enumerate(kmeans_model.labels_):
    plt.plot(x1[i], x2[i], color=colors[l], marker=markers[l],ls='None')
    plt.xlim([0, 10])
    plt.ylim([0, 10])

plt.scatter(centers[:,0], centers[:,1], marker="x", color='r')
plt.show()
```
